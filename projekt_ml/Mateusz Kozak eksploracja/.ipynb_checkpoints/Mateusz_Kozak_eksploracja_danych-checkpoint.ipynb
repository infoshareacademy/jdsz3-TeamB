```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline, FeatureUnion

df = pd.read_csv(r"Credit.csv")
df = df.drop(columns=["Unnamed: 0"])
df.head(5)
```

```python
numeric_columns = df.select_dtypes(include =[np.number]).columns.tolist()
cat_columns = df.select_dtypes(exclude=[np.number]).columns.tolist()
cat_columns
```

## Pair plot

```python
sb.pairplot(df[numeric_columns])
plt.show()
```

## Corelation Plot

```python
correlation = df[numeric_columns].corr()
plt.figure(figsize=(10,10))
sb.heatmap(correlation, cmap="RdBu_r")
plt.show()
```

## Box plots

```python

col = list(df[numeric_columns].columns)
print(col)

plt.figure(figsize=(10,10))
for i in range(7):
    plt.subplot(3,3,i+1)
    plt.boxplot(df[col[i]], notch=True)
    plt.title(col[i])
plt.show()
```

```python
from sklearn.manifold import TSNE

scaller = StandardScaler()
scalled_data = scaller.fit_transform(df[numeric_columns])

tsne = TSNE(n_components=2, perplexity=15, random_state=2).fit_transform(scalled_data)
plt.figure(figsize=(15,10))
sb.scatterplot(x=tsne.T[0], 
               y=tsne.T[1], 
               data=df[numeric_columns], 
               hue=df["Rating"], 
               palette="gist_rainbow_r",
               size=df["Income"]
               )
```

```python
scaller = MinMaxScaler()
scalled_data = scaller.fit_transform(df[numeric_columns])

tsne = TSNE(n_components=2, perplexity=10, random_state=2).fit_transform(scalled_data)
plt.figure(figsize=(15,10))
sb.scatterplot(x=tsne.T[0], 
               y=tsne.T[1], 
               data=df[numeric_columns], 
               hue=df["Rating"], 
               palette="gist_rainbow_r",
               size=df["Income"]
               )
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=2).fit_transform(scalled_data)
plt.figure(figsize=(15,10))
sb.scatterplot(x=pca.T[0], 
               y=pca.T[1], 
               data=df[numeric_columns], 
               hue=df["Rating"], 
               palette="gist_rainbow_r",
               size=df["Income"]
               )
```

```python
num_components = 5
pca = PCA(n_components=num_components, random_state=2)
transformed_data = pca.fit_transform(scalled_data)
data_variance = pca.explained_variance_
print(data_variance)

data_variance2 = pca.explained_variance_ratio_
print(f"variance ratio: {data_variance2}")

suma =0
a = 0
while a != num_components:
    suma += data_variance2[a]
    a+=1
print(f"suma wariancji: {suma}")
ex_variance = np.var(transformed_data, axis=0)
ex_variance_ratio = ex_variance*100/np.sum(ex_variance)
iterator_list = 0
cumul_variance =0

while iterator_list<num_components:
    cumul_variance+=ex_variance_ratio[iterator_list]
    iterator_list+=1
#print(cumul_variance)

#x = data_factors.get_covariance()
#print(x)

comp_labels =["PC"+str(i) for i in range(1, num_components+1)]
#print(comp_labels)

plt.figure(figsize=(15,15))
plt.subplot(2,1,1)
plt.bar(x=range(1,num_components+1), height=ex_variance_ratio, tick_label=comp_labels)
plt.ylabel("Percentage of variance")
plt.xlabel("Principal Component")
plt.title("Variance in terms of PC")

plt.subplot(2,1,2)
plt.plot(np.cumsum(data_variance))
plt.xlabel("Number of Components")
plt.ylabel("Variance Sum")
plt.title("Cumulative Variance")
plt.show()

```

```python
one = [1 for i in range(0,num_components+1)]

plt.figure(figsize=(10,10))
plt.plot(data_variance*100)
plt.plot(range(0,num_components+1), one, "r--")
plt.xlabel("Number of Components")
plt.ylabel("Variance")
plt.title("Variance")
plt.show()
```
